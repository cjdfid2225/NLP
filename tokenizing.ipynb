```
okt = Okt()
```

```
sentences = list(data['Q']) + list(data['A'])
```

```
sentences
```

```
# 문장을 형태소 별로 분석
morph_tokenized_sentences = [okt.morphs(s.replace(' ', '')) for s in sentences]
```

```
morph_tokenized_sentences[:3]
```

```
# 각 문장이 몇개의 형태소로 구성되어 있는지
sent_len_by_morph = [len(t) for t in morph_tokenized_sentences]
```

```
print(sent_len_by_morph)
```

```
tokenized_sentences = [s.split() for s in sentences]
```

```
sent_len_by_aeojeol = [len(token) for token in tokenized_sentences] #어절 개수
sent_len_by_eumjeol = [len(s.replace(' ','')) for s in sentences] #음절 개수
```

```
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))
plt.hist(sent_len_by_morph, bins=50, range=[0,50], alpha=0.5, color='blue', label='morph')
plt.hist(sent_len_by_aeojeol, bins=50, range=[0,50], alpha=0.5, color='yellow', label='aeojeol')
plt.hist(sent_len_by_eumjeol, bins=50, range=[0,50], alpha=0.5, color='red', label='eumjeol')

plt.title('Sentence Length Histogram')
plt.xlabel('Sentence Length')
plt.ylabel('Number of Sentence')
```

```
plt.boxplot([sent_len_by_eumjeol, sent_len_by_morph, sent_len_by_aeojeol],
            labels=['eumjeol','morph','aeojeol'],showmeans=True)
```

```
query_sentences = list(data['Q'])
ans_sentences = list(data['A'])

# 문장을 형태소 별로 분석
q_morph_tokenized_sentences = [okt.morphs(s.replace(' ', '')) for s in query_sentences]
a_morph_tokenized_sentences = [okt.morphs(s.replace(' ', '')) for s in ans_sentences]
```

```
q_sent_len_by_morph = [len(t) for t in q_morph_tokenized_sentences]
a_sent_len_by_morph = [len(t) for t in a_morph_tokenized_sentences]
```

```
plt.figure(figsize=(14,6))
plt.boxplot([q_sent_len_by_morph, a_sent_len_by_morph],
            labels=['Question', 'Answer'],
            showmeans=True)
```
